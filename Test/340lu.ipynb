```python
finbert = BertForSequenceClassification.from_pretrained('yiyanghkust/finbert-fls',num_labels=3)
tokenizer = BertTokenizer.from_pretrained('yiyanghkust/finbert-fls')

nlp = pipeline("text-classification", model=finbert, tokenizer=tokenizer, max_length=512, truncation=True)

df = pd.read_csv('reddit_finance_posts_cleaned.csv')

content = df.iloc[:, 3]
title = df.iloc[:, 1]

df['Content_Label'] = content.apply(lambda x: nlp(x)[0]['label'])
df['Title_Label'] = title.apply(lambda x: nlp(x)[0]['label'])

df.to_csv('reddit_finance_posts_with_FLSlabels.csv', index=False)

print("saved")
```

```python
#!pip install pyfin-sentiment

# Positive == Bullish
# Neutral == Uncertain
# Negative == Bearish

from pyfin_sentiment.model import SentimentModel

df = pd.read_csv('reddit_finance_posts_cleaned.csv')
SentimentModel.download("small")

content = df.iloc[:, 3].tolist()
title = df.iloc[:, 1].tolist()

model = SentimentModel("small")

contentPred = model.predict(content)
titlePred = model.predict(title)

df['Content_Label'] = contentPred
df['Title_Label'] = titlePred

df.to_csv('reddit_finance_posts_with_FinSentimentlabels.csv', index=False)

print("saved")
```
